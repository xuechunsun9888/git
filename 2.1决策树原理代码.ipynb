```python
from sklearn import tree # 决策树模块
from sklearn.datasets import load_wine # 红酒数据集
from sklearn.model_selection import train_test_split # 训练集、测试集的划分
import pandas as pd # Python里面的Excel表格
```

```python
# 导入红酒数据集
wine = load_wine()
```

```python
# 训练集测试集划分
Xtrain, Xtest, Ytrain, Ytest = train_test_split(wine.data, wine.target, test_size=0.3, random_state=7)
```

```python
# 测试集Y维度
Ytest.shape
```

```python
# 建模三步走
clf = tree.DecisionTreeClassifier(criterion="entropy") # 实例化一个clf对象，clf就是用来分类的
clf = clf.fit(Xtrain, Ytrain) # 训练集进行训练
score = clf.score(Xtest, Ytest) # 测试集进行测试

score
```

```python
# 决策树画图代码

feature_name = ['酒精','苹果酸','灰','灰的碱性','镁','总酚','类黄酮','非黄烷类酚类','花青素','颜色强度','色调','od280/od315稀释葡萄酒','脯氨酸']

import graphviz

dot_data = tree.export_graphviz(clf
                                ,feature_names=feature_name
                                ,class_names=["wine1", "wine2", "wine3"]
                                ,filled=True
                                ,rounded=True
                                )

graph = graphviz.Source(dot_data)
graph
```

```python
# 决策树具有随机性
clf = tree.DecisionTreeClassifier(criterion="entropy") # 实例化一个clf对象，clf就是用来分类的
clf = clf.fit(Xtrain, Ytrain) # 训练集进行训练
score = clf.score(Xtest, Ytest) # 测试集进行测试

score
```

```python
# 决策树random_state和splitter参数调整
clf = tree.DecisionTreeClassifier(criterion="entropy"
                                  ,random_state=30
                                  ,splitter="random"
                                  ) # 实例化一个clf对象，clf就是用来分类的
clf = clf.fit(Xtrain, Ytrain) # 训练集进行训练
score = clf.score(Xtest, Ytest) # 测试集进行测试

score
```

```python
# 画图
dot_data = tree.export_graphviz(clf
                                ,feature_names=feature_name
                                ,class_names=["wine1", "wine2", "wine3"]
                                ,filled=True
                                ,rounded=True
                                )

graph = graphviz.Source(dot_data)
graph
```

```python
# 树对训练集的拟合程度如何
score_train = clf.score(Xtrain, Ytrain)
score_train
```

```python
# 剪枝参数max_depth、min_samples_leaf、min_samples_split
clf = tree.DecisionTreeClassifier(criterion="entropy"
                                  ,random_state=30
                                  ,splitter="random"
                                  ,max_depth=3
                                  ,min_samples_leaf=5
                                  ,min_samples_split=30
                                  ) # 实例化一个clf对象，clf就是用来分类的
clf = clf.fit(Xtrain, Ytrain) # 训练集进行训练
score = clf.score(Xtest, Ytest) # 测试集进行测试

score

dot_data = tree.export_graphviz(clf
                                ,feature_names=feature_name
                                ,class_names=["wine1", "wine2", "wine3"]
                                ,filled=True
                                ,rounded=True
                                )

graph = graphviz.Source(dot_data)
graph
```

```python
# apply返回测试样本所在叶子节点的索引
clf.apply(Xtest)
```

```python
# predict返回每个测试样本的分类/回归结果
clf.predict(Xtest)
```

```python
# 绘制学习曲线

import matplotlib.pyplot as plt
test = []
for i in range(20):
    clf = tree.DecisionTreeClassifier(max_depth=i+1
                                      ,criterion="entropy"
                                      ,random_state=30
                                    #   ,splitter="random"
                                      ) # 建立深度不同的10颗树
    clf = clf.fit(Xtrain, Ytrain) # 训练集进行训练
    score = clf.score(Xtest, Ytest) # 测试集进行测试，得到分数accuracy，准确度
    test.append(score)

plt.plot(range(1,21), test, color="red", label="max_depth")
plt.legend()
plt.show()

```

```python
# 加利福尼亚房价数据集演示回归树
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeRegressor


housing = fetch_california_housing()
```

```python
housing
```

```python
housing.data
```

```python
housing.target
```

```python
housing.target.min()
```

```python
housing.target.max()
```

```python
# 建立回归树三步走
# 决策树回归器，评价方式MSE
regressor = DecisionTreeRegressor(random_state=0) # 实例化
cross_val_score(regressor, housing.data, housing.target, cv=10, # 交叉验证评价方式，10折交叉验证
                scoring = "neg_mean_squared_error" # 评价指标
               ).mean()
```

```python
# 模型评价指标类型
import sklearn
sklearn.metrics.get_scorer_names()
```

```python
# 决策树回归器，评价方式MAE
regressor = DecisionTreeRegressor(random_state=0) # 实例化
cross_val_score(regressor, housing.data, housing.target, cv=10,
                scoring = "neg_mean_absolute_error"
               ).mean()
```
